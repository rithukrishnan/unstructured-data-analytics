```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#scrapped data is stored in beer_reviews.csv file
data = pd.read_csv("PA_sen_analysis_data.csv")
```

```python
data
```

```python
data = data[["name", "party", "eval_score"]]
data
```

```python
data.reset_index()

col =[]
  
        
for index, row in data.iterrows():
    if row['party'] == "R":
        col.append("red")
    else:
        col.append("blue")
    
```

```python
fig, ax = plt.subplots(figsize=(30, 15))

for index, row in data.iterrows():
      
    # plotting the corresponding x with y 
    # and respective color
    plt.scatter(row["name"], row["eval_score"], c = col[index], s = 100,
                linewidth = 0)

    
for index, row in data.iterrows():
#for i, txt in enumerate(n):
    ax.annotate(row["name"], (row["name"], row["eval_score"]))

fig.set_facecolor('#B8BF79')
ax.set_facecolor("#EDEEEC")
plt.xlabel('Senates', fontsize=20)
plt.ylabel('Support Score', fontsize=20)
plt.title('Support for Marijuana', fontsize=30)
plt.xticks(data["name"], weight="bold", rotation = 90, fontsize=15)
plt.show()
```
